```python
# Select python version of pyNA 
import os
os.environ["pyna_language"] = 'python'
from pyNA.pyna import pyna

import pdb
import numpy as np
import openmdao.api as om
import pandas as pd
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import clear_output
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# Load default pyna settings
pyna_settings = pyna.load_settings(case_name = 'NASA STCA Standard')
pyna_settings.save_results = True
```

### Compute noise time series for predefined trajectory

```python
py = pyna(settings=pyna_settings)
py.compute_noise_time_series()
```

```python
py.plot_noise_time_series(metric='pnlt')
```

### Compute noise source distribution

```python
py = pyna(settings=pyna_settings)
py.settings.validation=False
py.plot_noise_source_distribution(time_step=152, metric='spl', components=['core', 'jet_mixing', 'airframe', 'fan_inlet', 'fan_discharge'])
```

### Compute epnl table for different noise sources

```python
py = pyna(settings=pyna_settings)
py.compute_noise_epnl_table()
```

### Compute noise contours

```python
# Compute noise contours
py = pyna(settings=pyna_settings)
py.compute_noise_contours(x_lst=np.linspace(0, 15000, 5), y_lst=np.linspace(0, 4000, 2))
```

```python
py.plot_noise_contours()
```
